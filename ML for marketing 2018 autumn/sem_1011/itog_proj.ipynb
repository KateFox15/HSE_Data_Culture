# Итоговый проект (как бы страшно это не звучало) 

Во всём курсе на написание кода были только задания на Datacamp. Пришло время исправить это недоразумение и заняться итоговым проектом.  Кекс был призван продемонстрировать то, насколько вы в состоянии работать со смысловой составляющей. Итоговый проект будет направлен на техническую составляющую.

Это задание будет довольно объёмным. В итоговой оценке оно занимает большую долю. Поэтому дерзайте! :) 

## Что надо делать?

Для решения вам предоставляется несколько задач. Выберете одну из них на свой вкус и цвет. Вам предстоит изучить представленные данные, понять как они устроены и построить модель, которая могла бы помочь решить поставленную задачу. 

Ваше исследование должно иметь прмерно следущую структуру: 

* Формулирование бизнес-задачи. Её перевод на язык машинного обучение, выбор и обоснование метрик качества. 
* Описание данных, их визуальный анализ. 
* Предобработка данных. Чётко опишите какую именно предобработку вам необходимо сделать, а главное поясните зачем. Попробуйте сварить в ходе предобработки новые объясняющие переменные, которые могут помочь в решении задачи.
* Выберете несколько моделей, которые могли бы вам помочь в решении поставленной задачи. Проведите моделирование.
* Постарайтесь достичь максимального качества при моделировании. 

Например, попробуйте перебрать параметры с помощью решётки, попробуйте несколько разных подходов. Сварите новые переменные, кластеризуйте те наблюдения, которыми вы располагаете и используйте кластеры в качестве новой объясняющей переменной (да, это легально). Не забудьте сделать OHE-кодирование... Можете попробовать отобрать только самые важные переменные с помощью Lasso-регрессии и уже на них обучить что-то другое. Можете делать всё, что хотите. Даже гонять нейросетки, если это правда поможет в решении задачи.

## Оценивание

Проект оценивается по 10-бальной шкале. Все шаги, которые были вами предприняты должны быть чётко аргументированы. За неаргументированные действия/результаты баллы ставится не будут. Следуйте структуре, описанной выше и всё у вас будет хорошо. Ваша работа будет оцениваться по двум составляющим: написанная вами тетрадка с кодом и выступление.
